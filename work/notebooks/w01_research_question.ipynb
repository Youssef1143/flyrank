```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
print("""
Question:
 Given a page's search/content signals over a prior 90-day window, can we predict whether it will decline or grow over the following 30 days?

Unit of analysis:
 one content page for one client, evaluated at a specific cutoff date (so the same page can appear multiple times if I use multiple cutoff dates).

Decision this improves:
 which pages a reviewer should proactively check before they visibly decline, rather than only reacting after impressions/clicks have already dropped.

Action a human takes:
 flagged pages go into a review queue for refresh, expansion, or monitoring — the same kind of action as Lane 2, but triggered earlier, based on predicted risk rather than an already-visible drop.

Cost of a false positive:
a reviewer spends time checking a page that was never actually going to decline — wasted but low-stakes time.

Cost of a false negative:
 a page that was quietly about to decline gets skipped, and by the time it's obviously declining (Lane 2's world), some of that traffic/opportunity is already lost — this is the more expensive error, since early warning is the entire value proposition of this direction.
""")
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

decline_share = (df["trend_direction"] == "down").mean()
eligible_share = ((df["impressions_90d"] > 0) & (df["content_age_days"] >= 90)).mean()

print("Rows:", len(df))
print("Share currently declining:", round(decline_share, 3))
print("Share eligible for a 90-day window:", round(eligible_share, 3))
```

To check whether this direction is worth pursuing, I loaded the starter dataset and looked at two things: how unstable the content already is, and whether pages have enough history to support a 90-day feature window.

30,000 rows in the starter dataset — a reasonable size to sanity-check ideas before moving to the full warehouse.
54.2% of pages are currently flagged as declining (trend_direction == "down"). This means instability is common, not rare — a real, sizeable share of content is worth prioritizing, which justifies building a system to catch this earlier.
100% of pages are eligible for a 90-day window (they meet the impressions_90d > 0 and content_age_days >= 90 filters). This confirms the starter data has enough history depth to support the prior-90-day / next-30-day window structure my direction depends on.

Together these numbers show the underlying instability is common enough to matter, and there's enough historical depth in the data to build the future-window labels my approach needs — supporting that this is worth the next 7 weeks, rather than a rare or thin problem.
